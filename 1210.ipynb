```python
from pingouin import pairwise_gameshowell
import scipy.stats as stats

def f_test(var):
    f_statistic, p_value = stats.f_oneway(data[data['ranking'] == 'Top_10'][var],
                                        data[data['ranking'] == 'remain'][var])

    print("ANOVA 결과:")
    print("F-statistic:", f_statistic.round(3))
    print("p-value:", p_value)

    posthoc = pairwise_gameshowell(data=data, dv=var, between='ranking')
    print("\nGames-Howell 사후 검증:")
    display(pd.DataFrame(posthoc).round(3))
    return None
```

```python
agg = data.drop(['date', 'game_size', 'match_id', 'match_mode', 'party_size', 'player_name', 'team_id', 'team_placement'], axis= 1)
agg.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# 데이터 스케일링(표준화 mu = 0, sigma = 1)
scaler = StandardScaler()
scaled_agg = scaler.fit_transform(agg)

```

```python
from sklearn.cluster import MiniBatchKMeans

kmeans = MiniBatchKMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(scaled_agg)
agg['Cluster'] = clusters
agg['team_placement'] = data['team_placement']
print(agg)
agg.groupby(by='Cluster').mean()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics import pairwise_distances_argmin_min
from tqdm import tqdm

# 미니배치 K-Means를 사용하여 엘보우 메서드로 적절한 클러스터 개수 찾기
distortions = []
cluster_range = range(1, 11)  # 클러스터 개수의 범위를 지정

for n_clusters in tqdm(cluster_range):
    kmeans = MiniBatchKMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(scaled_agg)
    distortions.append(sum(pairwise_distances_argmin_min(scaled_agg, kmeans.cluster_centers_)[1]))

# 엘보우 메서드 그래프 그리기
plt.figure(figsize=(8, 6))
plt.plot(cluster_range, distortions, marker='o')
plt.title('Elbow Method for Optimal Cluster Number')
plt.xlabel('Number of Clusters')
plt.ylabel('Distortion (Sum of Squared Distances)')
plt.grid(True)
plt.show()
```
